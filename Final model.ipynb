```python
df['fetal_health'] = df['fetal_health'].astype(int)
```

```python
df.describe()
```

```python
df.drop('severe_decelerations',axis=1,inplace=True)
```

```python
x=df.drop(columns='fetal_health')
y=df['fetal_health']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=35)
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(x_train)

# transform train and test sets
X_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
x_train
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=x_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=x_test.columns)
```

```python
np.round(x_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix,accuracy_score


knn= KNeighborsClassifier()
knn_scaled= KNeighborsClassifier()

tree= DecisionTreeClassifier()
tree_scaled= DecisionTreeClassifier()

sv= SVC()
sv_scaled= SVC()

rf= RandomForestClassifier()
rf_scaled= RandomForestClassifier()

grd= GradientBoostingClassifier()
grd_scaled= GradientBoostingClassifier()

```

# GridSearchCV

```python
n_estimators = [20,60,100,120]

max_features = [0.2,0.6,1.0]

max_depth = [2,8,None]

max_samples = [0.5,0.75,1.0]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
              'max_samples':max_samples
             }
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(estimator = rf, 
                       param_grid = param_grid, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)

rf_scaled_grid = GridSearchCV(estimator = rf, 
                       param_grid = param_grid, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
rf_grid.fit(x_train,y_train)

```

```python
rf_scaled_grid.fit(X_train_scaled,y_train)
```

```python
rf_scaled_grid.best_params_
```

```python
rf_scaled_grid.best_score_
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```

```python
rf.fit(x_train, y_train) 
rf_scaled.fit(X_train_scaled,y_train)

y_pred=rf.predict(x_test)
y_pred_scaled = rf_scaled.predict(X_test_scaled)

print("Actual Random Forst : ",accuracy_score(y_test,y_pred))
print("Scaled Random Forst : ",accuracy_score(y_test,y_pred_scaled))


```

# Decission Tree


```python
max_features = ['auto', 'sqrt','log2']
max_depth = [int(x) for x in np.linspace(10, 200,10)]
min_samples_split = [2, 5, 10,14]


splitter = ['best','random']
grid_cv = {    'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'splitter':splitter,
              'criterion':['entropy','gini']}
```

```python
tree_grid = GridSearchCV(estimator = tree, 
                       param_grid = grid_cv, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
tree_grid.fit(x_train,y_train)
```

```python
tree_grid.best_params_
```

```python
tree_grid.best_score_
```
